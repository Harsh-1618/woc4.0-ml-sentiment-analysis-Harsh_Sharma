### 2) Extracting numbers from string:

```python
extract = pd.Series(data['Unique ID'].copy(), name = "New Index")

# Alternate way
# extract = data['Unique ID'].copy()
# extract = pd.Series(extract, name = "New Index")

for row in range(len(extract)):
    temp = extract.loc[row]
    temp = list(temp)
    temp.remove('I')
    temp.remove('D')
    temp.remove('_')
    temp = "".join(temp)
    extract.loc[row] = temp
```

```python
extract = extract.astype('int64')
```

```python
# ---- data['Unique ID'].str.extract('(\d+)') ---- another method using regular expressions.
```

### 3) Indexing:

```python
data.set_index(extract, inplace = True)
data
```

### 4) Label Encoding:

```python
print(data['Regionname'].unique().shape)
print(data['Suburb'].unique().shape)

# Since "Regionname" column has less unique values, i am using One-Hot Encoding method for it, and since "Suburb" has many
# unique values i am using Label Encoding method for it.
```

```python
# Encoding for 'Suburb' column:
labelencoder = LabelEncoder()
data['Suburb_N'] = labelencoder.fit_transform(data['Suburb'])

# ---- data.drop(['Suburb'], inplace = True, axis = 1) ---- if you wanna get rid of categorical data.
data
```

```python
# Encoding for 'Regionname' column:

updated_index = list(range(1, len(data.index) + 1))  # cause we updated the index of 'data' dataframe in the first place

onehotencoder = OneHotEncoder(sparse = False, handle_unknown = "error", drop = "first")
ohe_df = pd.DataFrame(onehotencoder.fit_transform(data[['Regionname']]), index = updated_index)
ohe_df # 7 colunms instead of 8 cause the one which is dropped is represented by all zero values.
```

```python
data = data.join(ohe_df)

# ---- data.drop(['Regioname'], inplace = True, axis = 1) ---- if you wanna get rid of categorical data.

data
```

### 5) Dealing with NaN Values:

```python
data.dropna(how = 'all', axis = 0, inplace = True) # drop that row which is all NaN and hence useless
data.dropna(how = 'all', axis = 1, inplace = True) # drop that column which is all NaN and hence useless

print(data.isnull().sum())  # True is 1 and False is 0, so the sum() will give the number of NaN values in each column.
# so there are 62 NaN values in 'Car' column

data
```

```python
# CouncilArea is categorical data so we cannot interpolate it, rest we can. 

data['Car'] = data['Car'].interpolate()
data['BuildingArea'] = data['BuildingArea'].interpolate()  # apparantly it won't interpolate the first row
data['YearBuilt'] = data['YearBuilt'].interpolate()   # apparantly it won't interpolate the first row
data.dropna(subset=['CouncilArea'], how = 'any', axis = 0, inplace = True) # dropped all the rows where CouncilArea is NaN.

print(data.isnull().sum())
data
```
